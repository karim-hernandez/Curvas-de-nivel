```python
from ipywidgets import interact, fixed
import numpy as np
import matplotlib.pyplot as plt
def plot_xyz_product(f_txt, z_val, fix_zlimits= False):
    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)
    f = eval(f'lambda x, y, z : {f_txt}')
    Z = f(X, Y, z_val)
    plt.pcolormesh(X, Y, Z, shading='auto', cmap = 'jet', vmax = 1000 if fix_zlimits else Z.max(), vmin = -1000 if fix_zlimits else  Z.min())
    plt.colorbar()
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Plot of {f_txt}')
    plt.show()
f = lambda x, y, z : x*y*z
interact(plot_xyz_product, f_txt = 'x*y*z', z_val = (-10, 10, 0.01), fix_zlimits = False)
```

Lo que vemos a continuacion es un corte transversal de la funcion en el espacio a una coordenada $z = z_0$. el gradiente de color esta normalizado a que el limmite inferior sea -100 y el superior sea 1000

```python
def plot_3d(f, num_puntos = 100, c_value=0, tol = .1, z0 = 0, x0 = 0, y0 = 0):
  f = eval(f'lambda x, y, z :  {f}')
  delta = 1/num_puntos
  u, v, w = np.mgrid[-10:10:delta, -10:10:delta,  -10:10:delta]
  x, y = np.mgrid[-10:10:delta, -10:10:delta]
  colors = f(u, v, w)
  fig = plt.figure()
  ax = fig.add_subplot(231, projection='3d')

  scater = ax.scatter3D(u, v, w, c = colors, cmap = 'jet')
  fig.colorbar(scater)
  filter = abs(colors - c_value)<tol
  ax2 = fig.add_subplot(233, projection='3d')
  scater2 = ax2.scatter3D(u*filter, v*filter, w*filter, c = u*0 +c_value, cmap = 'jet', vmax =colors.max(), vmin =colors.min() )
  ax3 = fig.add_subplot(234)
  ax3.contourf(x, y, f(x0, x, y), cmap = 'jet', vmax =colors.max(), vmin =colors.min() )
  ax3.set_title('corte en x0')
  ax4 = fig.add_subplot(235)
  ax4.contourf(x, y, f(x, y0, y), cmap = 'jet', vmax =colors.max(), vmin =colors.min() )
  ax4.set_title('corte en y0')
  ax5 = fig.add_subplot(236)
  ax5.contourf(x, y, f(x, y, z0), cmap = 'jet', vmax =colors.max(), vmin =colors.min() )
  ax5.set_title('corte en z0')
  for axs in [ax, ax2, ax3, ax4, ax5]:
    axs.set_xlim(-10, 10)
    axs.set_ylim(-10, 10)
    try: axs.set_zlim(-10, 10)
    except: pass
  plt.show()
```

```python
interact(plot_3d, f = 'x**2+y**2+z**2', num_puntos = (1, 1001, 10), c_value = (-100,100, 1), tol = (0, 2, .1), x0 = (-10, 10, .1), y0 = (-10, 10, .1), z0 = (-10, 10, .1))
```
